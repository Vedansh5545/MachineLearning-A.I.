```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.linspace(0.1, 20 * np.pi, 20)
y = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```
